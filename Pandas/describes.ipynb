```python
from PDconfig import df_ML as MeLi, df_LV as LaVoz
from archivosDF import *

df_ML = abrirDF(MeLi)
df_LV = abrirDF(LaVoz)

df_LV.drop_duplicates(subset='id', keep='first', inplace=True, ignore_index=False)
df_ML.drop_duplicates(subset='id', keep='first', inplace=True, ignore_index=False)

df_LV = df_LV.dropna(how="all", subset=['terrenoTotal', 'terrenoEdificado', 'cantDormitorios', 'cantBanos', 'cantCochera'])
df_ML = df_ML.dropna(how="all", subset=['terrenoTotal', 'terrenoEdificado', 'cantDormitorios', 'cantBanos', 'cantCochera'])
```

```python
filterLC = ((df_LV["tipoPropiedad"] == "CASA") 
            & (df_LV["precioUSD"].between(df_LV["precioUSD"].quantile(0.1), 
                                        df_LV["precioUSD"].quantile(0.9)))
            & (df_LV["terrenoTotal"].between(df_LV["terrenoTotal"].quantile(0.1), 
                                        df_LV["terrenoTotal"].quantile(0.9)))
            & (df_LV["terrenoEdificado"].between(df_LV["terrenoEdificado"].quantile(0.1), 
                                        df_LV["terrenoEdificado"].quantile(0.9))))
casasL = df_LV.loc[filterLC]
casasL.describe()
```
